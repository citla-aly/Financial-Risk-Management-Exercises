```python
def meanLessStdSqrt(x):
    try:
        return (np.mean(x) - np.std(x))**2
    except:
        return np.nan
    
def sumChange(x):
    if len(x)>1:
        return np.diff(x).sum()
    else:
        return 0
    
```

```python
ing=df['Cierre'].rolling(2).agg(['mean','max',meanLessStdSqrt,sumChange])
```

```python
X=df[[x for x in df.columns if x not in ['Cierre','tgt','Adj Close','Fecha']]]
X
```

```python
y=df[['tgt']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mm=MinMaxScaler()
mm_y=MinMaxScaler()

xt,xv,yt,yv= train_test_split(X,y,train_size=.80,shuffle=False)
```

```python
xt=pd.DataFrame(mm.fit_transform(xt),columns=X.columns)

yt=pd.DataFrame(mm_y.fit_transform(yt),columns=['tgt'])

xv=pd.DataFrame(mm.transform(xv),columns=X.columns)
yv=pd.DataFrame(mm_y.transform(yv),columns=['tgt'])
```

```python
from sklearn.tree import DecisionTreeRegressor

from sklearn.linear_model import ElasticNet

from sklearn.svm import SVR
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
def metricas(y_true,y_pred):
    r2=r2_score(y_true,y_pred)
    mae=mean_absolute_error(y_true,y_pred)
    mse=mean_squared_error(y_true,y_pred)
    print(f'El r2 score es {r2}')
    print(f'El error cuadrático medio es {mse}')
    print(f'El error medio absoluto es {mae}')
```

```python
metricas(yv.iloc[:-1],tree.predict(xv.iloc[:-1]))
```

```python
metricas(yv.iloc[:-1],svm.predict(xv.iloc[:-1]))
```

```python
metricas(yv.iloc[:-1],elastic.predict(xv.iloc[:-1]))
```

```python
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
```

```python
param_elastic={'alpha':[x*.01 for x in range(1,100)], 'l1_ratio' : [x*.01 for x in range(0,100)]}
```

```python
Random_elastic=RandomizedSearchCV(ElasticNet(),param_elastic,scoring='r2',n_iter=1000,cv=3,verbose=1,n_jobs=-1,random_state=99)
```

```python
Random_elastic.fit(xt,yt)
```

```python
Random_elastic.best_score_
```

```python
Random_elastic.best_params_
```

```python
best_elastic=Random_elastic.best_estimator_
```

```python
y_estimada_elastic=best_elastic.predict(xv)

y_predict= pd.DataFrame(yv.values,columns=['real'])

y_predict['Estimada_elastic']=y_estimada_elastic
```
